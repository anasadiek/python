```python
import pandas as pd 
adverse = pd.read_csv('adverse_reactions.csv')
pat =  pd.read_csv('patients.csv')
treat = pd.read_csv('treatments.csv')
```

```python
pat['birthdate']   = pd.to_datetime(pat['birthdate'] )
```

```python
import datetime as dt 
pat['age']  =  2025 -  pat['birthdate'].dt.year


```

```python
pat
```

```python
#What’s the age distribution of patients? 
from matplotlib import pyplot as plt 
import seaborn as sns 
plt.figure(figsize=[15,9])
sns.histplot(data=pat,x='age')
```

```python
pat['given_name'] =  pat['given_name'].str.lower()
pat['surname'] =  pat['surname'].str.lower()

adverse['given_name'] =  adverse['given_name'].str.lower()
adverse['surname'] =  adverse['surname'].str.lower()

treat['given_name'] =  treat['given_name'].str.lower()
treat['surname'] =  treat['surname'].str.lower()
```

```python
pat['given_name'].str.lower()
```

```python
pat
```

```python
newpat =   pd.merge( left=pat,right=adverse , left_on=['given_name','surname'] , right_on=['given_name','surname']   )
```

```python
pat['age'].value_counts().sort_index()
```

```python
newpat['group_age']  = 0
newpat.loc[  (newpat['age'] >0)  &   (newpat['age'] <=50)  , 'group_age'    ]  = 'youth'
newpat.loc[  (newpat['age'] >50)  &   (newpat['age'] <=105)  , 'group_age'    ]  = 'Old'
```

```python
newpat
```

```python
plt.figure(figsize=[15,9])
sns.barplot(data=newpat, x='adverse_reaction',y= 'group_age',hue='assigned_sex')
```

```python
#Average patient age by sex.
plt.figure(figsize=[15,9])
sns.boxplot(data=newpat, x='assigned_sex',y='age')
```

```python
#Average patient age by sex.
newpat.groupby('assigned_sex').agg({'age':'mean'})
```

```python
pat = pd.merge(left= pat ,right= treat ,  left_on=['given_name','surname'] ,  right_on= ['given_name','surname']   )
```

```python
pat =  pat.drop(columns='country')
```

```python
pat['hba1c_change']  =   pat['hba1c_start'] - pat['hba1c_end']
```

```python
pat
```

```python
pat =  pd.melt(frame=pat,  value_name='treat_value'   ,   value_vars=['auralin','novodra']    ,var_name= 'treat_type',id_vars= ['patient_id',
 'assigned_sex',
 'given_name',
 'surname',
 'address',
 'city',
 'state',
 'zip_code',
 'contact',
 'birthdate',
 'weight',
 'height',
 'bmi',
 'age',
 'group_age',
 'hba1c_start',
 'hba1c_end',
 'hba1c_change'])
```

```python
pat.columns.to_list()
```

```python
pat
```

```python
import numpy as np 
pat.loc[ pat['treat_value'] == '-' , 'treat_value'  ] = np.nan
```

```python
pat.isna().sum()
```

```python
pat  =  pat.dropna()
```

```python
pat
```

```python
#Which drug (Auralin or Novodra) results in better HBA1C improvement?
plt.figure(figsize=[22,11])
sns.barplot(data=pat,x='treat_type',y='hba1c_change',estimator='mean')
```

```python
#Which drug (Auralin or Novodra) results in better HBA1C improvement?
pat.groupby(['treat_type']).agg({'hba1c_change':'mean'})
```

```python
#Are there patterns between demographics and treatment success?

pat.groupby(['state']).agg({'hba1c_change':'mean'})


```

```python
plt.figure(figsize=[27,11])
sns.barplot(data=pat,x='state',y='hba1c_change',estimator='mean')
```

```python
#Average HBA1C change per treatment type.
plt.figure(figsize=[22,11])
sns.barplot(data=pat,x='treat_type',y='hba1c_change',estimator='mean')

pat.groupby(['treat_type']).agg({'hba1c_change':'mean'})
```
