```python
import pandas as pd 
file =  pd.read_csv('gender_classification_v7.csv')
```

```python
file
```

```python
from sklearn.preprocessing import LabelEncoder
file['gender'] =  LabelEncoder().fit_transform(file['gender'])
```

```python
x =  file.drop(columns={'gender'})
y =  file['gender']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest =  train_test_split(x,y,test_size=.80,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrainscaled = scaler.fit_transform(xtrain)
xtestscaled = scaler.transform(xtest)

```

```python
from sklearn.linear_model import LogisticRegression

logmodel =  LogisticRegression()

logmodel.fit(xtrainscaled,ytrain)
```

```python
logmodelpredict =  logmodel.predict(xtestscaled)
```

```python
import plotly.express as px 

fig =  px.scatter(x=logmodelpredict,y=ytest)
fig.show()
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(ytest,logmodelpredict)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnmodel =  KNeighborsClassifier(n_neighbors=21)
knnmodel.fit(xtrainscaled,ytrain)
```

```python
knnpredict =  knnmodel.predict(xtestscaled)
```

```python
accuracy_score(ytest,knnpredict)
```

```python
from sklearn.naive_bayes import GaussianNB
naivemodel =  GaussianNB()
naivemodel.fit(xtrainscaled,ytrain)
```

```python
naivepredict =  naivemodel.predict(xtestscaled)
```

```python
accuracy_score(ytest,naivepredict)
```
