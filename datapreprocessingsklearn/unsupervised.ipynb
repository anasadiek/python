# Power Score Prediction 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
from sklearn.preprocessing import MinMaxScaler



linearmodel =  LinearRegression()
svr =   SVR()
knnr =   KNeighborsRegressor()
dtr =    DecisionTreeRegressor()
scaler =  MinMaxScaler()
encoder =  LabelEncoder()
```

```python
fileencoded = file.copy()
fileencoded = fileencoded.drop(columns={'Name','Name'})
fileencoded['Name'] = encoder.fit_transform(file['Name'])
fileencoded['Universe'] = encoder.fit_transform(file['Universe'])
fileencoded['Alignment'] = encoder.fit_transform(file['Alignment'])
fileencoded['Weapon'] = encoder.fit_transform(file['Weapon'])
```

```python
x= fileencoded.drop(columns={'Power Score','Name'})
y =  fileencoded['Power Score']

X=  scaler.fit_transform(x)
xtrain,xtest,ytain,ytest =  train_test_split(X,y,train_size=.80,random_state=18)

#xtrainscaled =  scaler.fit_transform(xtrain)
#xtestscaled =  scaler.transform(xtest)


```

```python
regmodels =  {'linearmodel':linearmodel,
              'svr':svr,
              'knnr':knnr,
              'dtr':dtr
              }

regresults  =  []

for name,model in regmodels.items():
    model.fit(xtrain,ytain)
    modelpredict =  model.predict(xtest)
    r2score =  r2_score(modelpredict,ytest)
    meanabsoluteerror =  mean_absolute_error(modelpredict,ytest)
    meansquareerror =  mean_squared_error(modelpredict,ytest)
    rootmeansquareerrorr  =  root_mean_squared_error(modelpredict,ytest)
    regresults.append( [name,r2score,meanabsoluteerror,meansquareerror,rootmeansquareerrorr]            )

```

```python
pd.DataFrame(regresults,columns=['name','r2score','mean_ABs','Mean_SQ','ROOT_Mean'] )
```

# Unsupervised Learn 

```python
from sklearn.cluster import KMeans
unfile =  file.copy()
unfile =  unfile.drop(columns={'Name'})
unfile['Universe'] =  encoder.fit_transform(unfile['Universe'])
unfile['Alignment'] =  encoder.fit_transform(unfile['Alignment'])
unfile['Weapon'] =  encoder.fit_transform(unfile['Weapon'])
```

```python
kmeanmodel =  KMeans(n_clusters=4,random_state=19)
kmeanpredict =  kmeanmodel.fit_predict(unfile)
```

```python
unfile['kmean_cluster'] =  kmeanpredict
```

```python
from sklearn.cluster import DBSCAN
dbmodel =  DBSCAN(eps=12,min_samples=6)
dbscanpredict =  dbmodel.fit_predict(unfile.drop(columns={'kmean_cluster'}))
dbscanpredict
```

```python
unfile['db_scan'] = dbscanpredict
```

```python
from sklearn.cluster import AgglomerativeClustering
heraric_model =  AgglomerativeClustering(n_clusters=4)
herricpredict =  heraric_model.fit_predict(unfile.drop(columns={'kmean_cluster','db_scan'}))
herricpredict
```

```python
unfile['heraric_clusters'] =  herricpredict
```
