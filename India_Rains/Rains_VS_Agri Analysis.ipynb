```python
import pandas as pd 

file =  pd.read_csv('rain-agriculture.csv')
# DataLink https://www.kaggle.com/datasets/naivedatamodel/india-rain-agriculture-over-the-span-of-52-years
```

```python
file.head()
```

# Dropping Production & Area Columns 

```python
file=  file.drop(columns=file.columns[file.columns.str.contains('AREA|PRODUCTION'  ).tolist()])
```

```python
# adding total year rain instead of quarters and will droppoing quarters for bettwe exploration 
file['total_rain'] =  file['AUG'] + file['JUL'] + file['JUN'] + file['SEP']
# renaming year agri and year of rain of better column selection 
file =  file.rename(columns={'Year':'Agri_year','YEAR':'rain_year'})
# dropping rain quarters 

file =  file.drop(columns={'JUN','JUL','AUG','SEP'})
```

```python
file.head()
```

# Rain Analysis vs Agri Analysis 


```python
# we group rain per year and yield per year for now we ignore per state we will analyze total india 
aggdata =  file.groupby(['rain_year']).agg({'RICE YIELD (Kg per ha)':'sum',
                                 'WHEAT YIELD (Kg per ha)':'sum',
                                 'KHARIF SORGHUM YIELD (Kg per ha)':'sum',
                                 'RABI SORGHUM YIELD (Kg per ha)':'sum',
                                 'SORGHUM YIELD (Kg per ha)':'sum',
                                 'PEARL MILLET YIELD (Kg per ha)':'sum',
                                 'MAIZE YIELD (Kg per ha)':'sum',
                                 'FINGER MILLET YIELD (Kg per ha)':'sum',
                                 'BARLEY YIELD (Kg per ha)':'sum',
                                 'CHICKPEA YIELD (Kg per ha)':'sum',
                                 'PIGEONPEA YIELD (Kg per ha)':'sum',
                                 'MINOR PULSES YIELD (Kg per ha)':'sum',
                                 'GROUNDNUT YIELD (Kg per ha)':'sum',
                                 'SESAMUM YIELD (Kg per ha)':'sum',
                                 'RAPESEED AND MUSTARD YIELD (Kg per ha)':'sum',
                                'SAFFLOWER YIELD (Kg per ha)':'sum',
                                'CASTOR YIELD (Kg per ha)':'sum',
                                'LINSEED YIELD (Kg per ha)':'sum',
                                'SUNFLOWER YIELD (Kg per ha)':'sum',
                                'SOYABEAN YIELD (Kg per ha)':'sum',
                                'OILSEEDS YIELD (Kg per ha)':'sum',
                                'SUGARCANE YIELD (Kg per ha)':'sum',
                                'COTTON YIELD (Kg per ha)':'sum',
                                'total_rain':'sum'
                                





                                 }).reset_index()
```

```python
import plotly.express as px 
# lets check our yeild how effected by totalrain
aggdata.corr()
graph =  px.imshow(aggdata.corr(),height=900)
graph.show()
pd.DataFrame(aggdata.corr()['total_rain'].sort_values(ascending=False))
```
